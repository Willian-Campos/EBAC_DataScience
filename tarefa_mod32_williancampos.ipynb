# Probabilidade - Tarefa

```python
import numpy as np
from scipy import stats
from scipy.stats import norm, t

import seaborn as sns
import matplotlib.pyplot as plt
```

1. Uma máquina embala ração para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatório, com distribuição Normal, com média 15kg, mas com desvio padrão 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produção tenha mais que 15,2 kg.

```python
Z = (15.2 - 15) / 0.1 #15.2 LIMITE-SUPERIOR  #15 MÉDIA  #0.1 DESVIO PADRÃO
print(f'Pontuação Z: {Z}')
probabilidade_acumulada = norm.cdf(Z)
print(f'Probabilidade Acumulada: {probabilidade_acumulada}')
maior_que_15_2 = 1 - probabilidade_acumulada
print(f'Probabilidade de um pacote ser maior que 15,2 sabendo do desvio padrão de 0,1: {maior_que_15_2}')
```

2. Da máquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
# Usando SciPy Norm
# 0.95 percentil, 15 média, 0.1 desvio padrão
percentil_95 = norm.ppf(0.95, loc=15, scale=0.1)
print(f'Percentil de 95%: {percentil_95}')
```

3. Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na célula abaixo:

```python
# -6 limite inferior, 6 limite superior, 200 é a quantidade de elementos
Y = np.linspace(-6, 6, 200)

# Y são os elementos das lista, 0 média e 1 é o desvio padrão (como é 1 não tem necessidade de tirar raiz)
Y1 = norm.cdf(Y, loc=0, scale=1)

# Y são os elementos da lista, 5 são os 5 graus de liberdade, t.cdf para identificar como t-studdent
Y2 = t.cdf(Y, df=5)


```

```python
# Resultado das probabilidades do Y1
for i in range(len(Y)):
    print(f"Probabilidade de Y1 ser menor que {Y[i]}: {Y1[i]}")
```

```python
# Resultado das probabilidades do Y2
for i in range(len(Y)):
    print(f"Probabilidade de Y1 ser menor que {Y[i]}: {Y2[i]}")
```

4. Faça uma visualização gráfica dos dados do item anterior.

```python
# Determinando tamanho do gráfico, largura e altura
plt.figure(figsize=(10, 6))

# Plot do Y1, com média 0 e desvio padrão 1 em AZUL
sns.lineplot(x=Y, y=Y1, label='Normal Padrão (média=0, desvio padrão=1)', color='blue')

# Plot da distribuição t-Student com 5 graus de liberdade em LARANJA
sns.lineplot(x=Y, y=Y2, label='t-Student (5 graus de liberdade)', color='orange')

# Adicionando legendas e título
plt.title('Probabilidade Acumulada das Distribuições Normal Padrão e t-Student')
plt.xlabel('Valores de Y')
plt.ylabel('Probabilidade Acumulada')

# Exibindo o gráfico
plt.legend()
plt.show()
```

5. 3. Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
# Valores de 0 a 1, 20 elementos
Y = np.linspace(0, 1, 20)

# Y1 Distribuição Normal (média 0 e variância 1)
percentis_Y1 = norm.ppf(Y, loc=0, scale=1)
print("Percentis para Y1 (Normal Padrão):", percentis_Y1)

print('\n\n') # Espaço entre as respostas

# Y2 t studdent com 5 graus de liberdade
percentis_Y2 = t.ppf(Y, df=5)
print("Percentis para Y2 (t-Student):", percentis_Y2)
```

6. Faça um gráfico com os dados do item anterior.

```python
# Tamanho da figura
plt.figure(figsize=(10, 6))

# Linhas do Y1 e Y2 com o conjunto Y
sns.lineplot(x=Y, y=percentis_Y1, label='Normal Padrão (Y1)')
sns.lineplot(x=Y, y=percentis_Y2, label='t-Student (Y2)')

plt.title('Percentis para Distribuições Normais Padrão e t-Student')
plt.xlabel('Y')
plt.ylabel('Percentil')

plt.show()
```

#### Conforme a quantidade de elementos aumentam ou diminuem o gráfico vai seguindo uma tendencia de formato
