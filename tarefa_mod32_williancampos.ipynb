# Probabilidade - Tarefa

```python
import numpy as np
from scipy import stats
from scipy.stats import norm, t

import seaborn as sns
import matplotlib.pyplot as plt
```

1. Uma mquina embala rao para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatrio, com distribuio Normal, com mdia 15kg, mas com desvio padro 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produo tenha mais que 15,2 kg.

```python
Z = (15.2 - 15) / 0.1 #15.2 LIMITE-SUPERIOR  #15 MDIA  #0.1 DESVIO PADRO
print(f'Pontuao Z: {Z}')
probabilidade_acumulada = norm.cdf(Z)
print(f'Probabilidade Acumulada: {probabilidade_acumulada}')
maior_que_15_2 = 1 - probabilidade_acumulada
print(f'Probabilidade de um pacote ser maior que 15,2 sabendo do desvio padro de 0,1: {maior_que_15_2}')
```

2. Da mquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
# Usando SciPy Norm
# 0.95 percentil, 15 mdia, 0.1 desvio padro
percentil_95 = norm.ppf(0.95, loc=15, scale=0.1)
print(f'Percentil de 95%: {percentil_95}')
```

3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na clula abaixo:

```python
# -6 limite inferior, 6 limite superior, 200  a quantidade de elementos
Y = np.linspace(-6, 6, 200)

# Y so os elementos das lista, 0 mdia e 1  o desvio padro (como  1 no tem necessidade de tirar raiz)
Y1 = norm.cdf(Y, loc=0, scale=1)

# Y so os elementos da lista, 5 so os 5 graus de liberdade, t.cdf para identificar como t-studdent
Y2 = t.cdf(Y, df=5)


```

```python
# Resultado das probabilidades do Y1
for i in range(len(Y)):
    print(f"Probabilidade de Y1 ser menor que {Y[i]}: {Y1[i]}")
```

```python
# Resultado das probabilidades do Y2
for i in range(len(Y)):
    print(f"Probabilidade de Y1 ser menor que {Y[i]}: {Y2[i]}")
```

4. Faa uma visualizao grfica dos dados do item anterior.

```python
# Determinando tamanho do grfico, largura e altura
plt.figure(figsize=(10, 6))

# Plot do Y1, com mdia 0 e desvio padro 1 em AZUL
sns.lineplot(x=Y, y=Y1, label='Normal Padro (mdia=0, desvio padro=1)', color='blue')

# Plot da distribuio t-Student com 5 graus de liberdade em LARANJA
sns.lineplot(x=Y, y=Y2, label='t-Student (5 graus de liberdade)', color='orange')

# Adicionando legendas e ttulo
plt.title('Probabilidade Acumulada das Distribuies Normal Padro e t-Student')
plt.xlabel('Valores de Y')
plt.ylabel('Probabilidade Acumulada')

# Exibindo o grfico
plt.legend()
plt.show()
```

5. 3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
# Valores de 0 a 1, 20 elementos
Y = np.linspace(0, 1, 20)

# Y1 Distribuio Normal (mdia 0 e varincia 1)
percentis_Y1 = norm.ppf(Y, loc=0, scale=1)
print("Percentis para Y1 (Normal Padro):", percentis_Y1)

print('\n\n') # Espao entre as respostas

# Y2 t studdent com 5 graus de liberdade
percentis_Y2 = t.ppf(Y, df=5)
print("Percentis para Y2 (t-Student):", percentis_Y2)
```

6. Faa um grfico com os dados do item anterior.

```python
# Tamanho da figura
plt.figure(figsize=(10, 6))

# Linhas do Y1 e Y2 com o conjunto Y
sns.lineplot(x=Y, y=percentis_Y1, label='Normal Padro (Y1)')
sns.lineplot(x=Y, y=percentis_Y2, label='t-Student (Y2)')

plt.title('Percentis para Distribuies Normais Padro e t-Student')
plt.xlabel('Y')
plt.ylabel('Percentil')

plt.show()
```

#### Conforme a quantidade de elementos aumentam ou diminuem o grfico vai seguindo uma tendencia de formato
