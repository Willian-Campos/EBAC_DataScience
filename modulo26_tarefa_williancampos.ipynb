# Tarefa - Álgebra Linear

A célula abaixo gera dados de notas de duas provas, p1 e p2, conforme em aula, mas com mais alunos.

## 1. Reduzindo correlações

a) Já vimos que expressar os dados através da média e diferença é uma transformação linear. Obtenha a média e a diferença através de uma multiplicação matricial, conforme visto em aula.

b) Calcule a correlação entre os dados da **p1** e **p2**

c) Calcule a correlação entre os dados transformados (média e diferença)

d) Comente as correlações observadas.

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python


%matplotlib inline

# Gerando os dados
np.random.seed(123)

p1 = np.random.random(50) *10
p1[p1>10] = 10
p1[p1<0] = 0

p2 = p1 + np.random.normal(0,1,50)
p2[p2>10] = 10
p2[p2<0] = 0

df = pd.DataFrame({'p1': p1, 'p2': p2})
```

```python
df
```

#### a) Já vimos que expressar os dados através da média e diferença é uma transformação linear. Obtenha a média e a diferença através de uma multiplicação matricial, conforme visto em aula.

```python
# Incluindo a coluna média
df['média'] = df.mean(axis=1)
df
```

```python
df['diferença'] = df.p2 - df.p1
df
```

#### b) Calcule a correlação entre os dados da **p1** e **p2**

##### Utilizando alguns outros meios antes de usar a multiplicação de matrizes

```python
correlation_matrix = df.corr()
correlation_matrix
```

##### Nessa visualização já da pra perceber que as notas tem uma boa correlação, ou seja, quem foi bem na primeira prova, normalmente vai bem na segunda prova e vice-versa, a correlação da média com as provas já era esperada, mas também tem a interessante informação da segunda prova com uma correlação maior.

```python
sns.set(style="white")
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Matriz de Correlação')
plt.show()
```

#### No gráfico abaixo foi feito uma transformação de troca de eixos, o P1 é direcionado nas posições do P2 e vice versa para comparar com os dados originais.

```python
# Definindo figura e eixos
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(1, 1, 1)

# Matriz de transformação
m = np.array([[0, 1],
              [1, 0]])

# Criando a matriz de transformação, multiplicando matrizes com o matmul e passando o m
df_rot = np.matmul(m, df[['p1', 'p2']].to_numpy().transpose())

# Plotando os dados originais
plt.scatter(df['p1'], df['p2'], label='Originais')

# Plotando os dados transformados
plt.scatter(df_rot[0, :], df_rot[1, :], label='Transformados')

# Adiciona setas representando os vetores da matriz de transformação
ax.arrow(0, 0, m[0, 0], m[1, 0], head_width=0.25, head_length=0.1, fc='g', ec='g', length_includes_head=True)
ax.arrow(0, 0, m[0, 1], m[1, 1], head_width=0.25, head_length=0.1, fc='r', ec='r', length_includes_head=True)

# Adiciona linhas horizontais e verticais de referência
ax.axvline(0, linewidth=0.5, linestyle='--')
ax.axhline(0, linewidth=0.5, linestyle='--')

# Configura limites dos eixos, escolhido 10+0.5 para passar um pouco da nota máxima
ax.set_xlim([-10.5, 10.5])
ax.set_ylim([-10.5, 10.5])

# Adicionar rótulos
plt.xlabel('p1')
plt.ylabel('p2')

# Adicionar legenda
plt.legend()

# Mostrar o gráfico
plt.show()
```

#### c) Calcule a correlação entre os dados transformados (média e diferença)

##### A média e diferença não tiveram muita correlação, foi alterado o angulo em 45 graus para ver uma perspectiva diferente, porém não pude extrair nenhuma informação util.

```python
# Definindo figura e eixos
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(1, 1, 1)

# Matriz de transformação
m = np.array([[0.707, -0.707],
              [0.707, 0.707]])

# Criando a matriz de transformação, multiplicando matrizes com o matmul e passando o m
df_rot = np.matmul(m, df[['média', 'diferença']].to_numpy().transpose())

# Plotando os dados originais
plt.scatter(df['média'], df['diferença'], label='Originais')

# Plotando os dados transformados
plt.scatter(df_rot[0, :], df_rot[1, :], label='Transformados')

# Adiciona setas representando os vetores da matriz de transformação
ax.arrow(0, 0, m[0, 0], m[1, 0], head_width=0.25, head_length=0.1, fc='g', ec='g', length_includes_head=True)
ax.arrow(0, 0, m[0, 1], m[1, 1], head_width=0.25, head_length=0.1, fc='r', ec='r', length_includes_head=True)

# Adiciona linhas horizontais e verticais de referência
ax.axvline(0, linewidth=0.5, linestyle='--')
ax.axhline(0, linewidth=0.5, linestyle='--')

# Configura limites dos eixos, escolhido 10+0.5 para passar um pouco da nota máxima
ax.set_xlim([-10.5, 10.5])
ax.set_ylim([-10.5, 10.5])

# Adicionar rótulos
plt.xlabel('média')
plt.ylabel('diferença')

# Adicionar legenda
plt.legend()

# Mostrar o gráfico
plt.show()
```

## 2. Soma e diferença

Realize agora uma transformação semelhante, mas que ao invés da média, dê a soma da **p1** e **p2**. Construa o gráfico e conclua que diferenças visuais você observa.

```python
df['soma'] = df.p1 + df.p2
df
```

```python
correlation_matrix = df.corr()
correlation_matrix
```

```python
# Definindo figura e eixos
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(1, 1, 1)

# Matriz de transformação
m = np.array([[0.407, -0.407],
              [0.407, 0.407]])

# Criando a matriz de transformação, multiplicando matrizes com o matmul e passando o m
df_rot = np.matmul(m, df[['soma', 'diferença']].to_numpy().transpose())

# Plotando os dados originais
plt.scatter(df['soma'], df['diferença'], label='Originais')

# Plotando os dados transformados
plt.scatter(df_rot[0, :], df_rot[1, :], label='Transformados')

# Adiciona setas representando os vetores da matriz de transformação
ax.arrow(0, 0, m[0, 0], m[1, 0], head_width=0.25, head_length=0.1, fc='g', ec='g', length_includes_head=True)
ax.arrow(0, 0, m[0, 1], m[1, 1], head_width=0.25, head_length=0.1, fc='r', ec='r', length_includes_head=True)

# Adiciona linhas horizontais e verticais de referência
ax.axvline(0, linewidth=0.5, linestyle='--')
ax.axhline(0, linewidth=0.5, linestyle='--')

# Configura limites dos eixos, escolhido 10+0.5 para passar um pouco da nota máxima
ax.set_xlim([-10.5, 10.5])
ax.set_ylim([-10.5, 10.5])

# Adicionar rótulos
plt.xlabel('soma')
plt.ylabel('diferença')

# Adicionar legenda
plt.legend()

# Mostrar o gráfico
plt.show()
```

##### O útimo gráfico foi usado o mesmo angulo por se tratar de uma comparação similar, como tanto a soma quanto a média usam das mesmas variáveis para ter seus resultados, ambos tiveram a exata mesma correlação, mas devido a natureza matemática a soma fez com que os pontos ficassem mais distântes uns dos outros, então se os vetores forem esticados isso pode dar uma maior diferença, então foi usado valores menores apesar de ser a mesma direção.
