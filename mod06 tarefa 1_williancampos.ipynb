####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
fig1, axs1 = plt.subplots(
    nrows=2, 
    ncols=2, 
    gridspec_kw={'hspace': 0.8}, 
    figsize=(12, 10))
sns.countplot(x='SEXO', data=df, ax=axs1[0, 0])

sns.countplot(x='GESTACAO', data=df, ax=axs1[0, 1])
axs1[0, 1].tick_params(axis='x', labelrotation=75)

sns.countplot(x='ESTCIVMAE', data=df, ax=axs1[1, 0])

sns.countplot(x='CONSULTAS', data=df, ax=axs1[1, 1])

```
